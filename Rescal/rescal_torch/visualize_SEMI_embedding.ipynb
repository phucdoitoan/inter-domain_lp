```python
import torch
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt

import pickle
```

```python
def visualize_embeddings(A1, A2, label_0=None, label_1=None, figname=None, xmin=-65, xmax=65, ymin=-65, ymax=65):
    A12 = torch.cat((A1, A2), dim=0)

    tsne = TSNE()

    A12_2d = tsne.fit_transform(A12)

    n1 = A1.shape[0]
    n2 = A2.shape[0]

    A1_2d = A12_2d[:n1]
    A2_2d = A12_2d[n1:]
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    axes[0].scatter(A1_2d[:,0], A1_2d[:,1], c=label_0, alpha=1, marker='.')
    axes[1].scatter(A2_2d[:,0], A2_2d[:,1], c=label_1, alpha=1, marker='.')
    
    axes[0].set_title('E1')
    axes[1].set_title('E2')
    
    axes[0].set_xlim(xmin, xmax)
    axes[1].set_xlim(xmin, xmax)
    
    axes[0].set_ylim(ymin, ymax)
    axes[1].set_ylim(ymin, ymax)
    
    if figname is not None:
        plt.savefig(figname)
```

```python
import pickle
```

```python
kg_dict = pickle.load(open('../data/FB15k-237/semi_3/divided/fix_intra-test/kg_valid_dict.pkl', 'rb'))
n1 = kg_dict['n1']
n_common = kg_dict['n_common']
n2 = kg_dict['n2']

print('n1: ', n1, 'n_common: ', n_common, 'n2: ', n2)
```

```python
path = 'fix-intra_semi_embeddings/'

rescal_emb = 'FB15k-237_WARM-START_SEMI-3_all_ent_embedding_alpha=0.0_lamda=1.0_semi=0.0_repeat=0.pt'
```

```python
# Load embeddings when training WITHOUT OT
semi_A = torch.load(path + rescal_emb).detach()
semi_A1 = semi_A[:n1+n_common]
semi_A2 = semi_A[n1:]
semi_A1.shape, semi_A2.shape
```

```python
visualize_embeddings(semi_A1, semi_A2, figname='emb_figures/FB15k-237-3_rescal.png')
```

```python
ot_emb = 'FB15k-237_WARM-START_SEMI-3_all_ent_embedding_alpha=0.6702509182507095_lamda=1.0_semi=0.0_repeat=0.pt'
```

```python
# Load embeddings when training WITH OT
ot_A  = torch.load(path+ot_emb).detach()
ot_A1 = ot_A[:n1+n_common]
ot_A2 = ot_A[n1:]
ot_A1.shape, ot_A2.shape
```

```python
visualize_embeddings(ot_A1, ot_A2, figname='emb_figures/FB15k-237-3_ot.png')
```
