```python
import torch
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt

import pickle

from sklearn.mixture import BayesianGaussianMixture
```

```python
def visualize_embeddings(A1, A2, label_0=None, label_1=None):
    A12 = torch.cat((A1, A2), dim=0)

    tsne = TSNE()

    A12_2d = tsne.fit_transform(A12)

    n1 = A1.shape[0]
    n2 = A2.shape[0]

    A1_2d = A12_2d[:n1]
    A2_2d = A12_2d[n1:]
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    axes[0].scatter(A1_2d[:,0], A1_2d[:,1], c=label_0, alpha=1, marker='.')
    axes[1].scatter(A2_2d[:,0], A2_2d[:,1], c=label_1, alpha=1, marker='.')
```

```python
# Load embeddings when training WITH OT
with open('stuff/ent_embedding_0_mul-ot_lamda=1.0.pt', 'rb') as file:
    A1_ot = torch.load(file)

with open('stuff/ent_embedding_1_mul-ot_lamda=1.0.pt', 'rb') as file:
    A2_ot = torch.load(file)
```

```python
# Visualize embeddings when training WITH OT: only wasserstein
# => make the absolute position become similar
# => the shapes of the two are similar and similar to the original
visualize_embeddings(A1_ot, A2_ot)
```

```python
# Load embeddings when training WITH OT
with open('stuff/ent_embedding_0_mul-ot_lamda=0.8.pt', 'rb') as file:
    A1_wgw = torch.load(file)

with open('stuff/ent_embedding_1_mul-ot_lamda=0.8.pt', 'rb') as file:
    A2_wgw = torch.load(file)
```

```python
# Visualize embeddings when training WITH OT: wgw lamda=0.8
# makes sense => both distributions' shapes and absolute positions are similar
visualize_embeddings(A1_wgw, A2_wgw)
```

```python
# Load embeddings when training WITHOUT OT
with open('stuff/ent_embedding_0.pt', 'rb') as file:
    A1 = torch.load(file)

with open('stuff/ent_embedding_1.pt', 'rb') as file:
    A2 = torch.load(file)
```

```python
# Visualize embeddings when training WITHOUT OT
# => the two distributions are similar
#    but not in a same absolute position
visualize_embeddings(A1, A2)
```

```python
def mrr(l):
    n = len(l)
    s = 0
    for u in l:
        s += 1/u
    
    return s/n


def mean_r(l):
    s = sum(l)
    return s/len(l)
```

```python
l = [1, 70, 100]
mrr(l), mean_r(l)
```

```python
l = [3, 4, 5]
mrr(l), mean_r(l)
```
